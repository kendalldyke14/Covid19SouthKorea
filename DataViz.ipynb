```python
def groupby_date(df, date_col, date_type, group_by = None):
    df[date_col] = pd.to_datetime(df[date_col])
    
    if group_by == None:
        df_month = df.set_index(date_col).resample(date_type).mean().reset_index()
    else:
        df_month = df.set_index(date_col).groupby(group_by).resample(date_type
                                                            ).mean().reset_index()
    
    return df_month
```

```python
search_month = groupby_date(search, "date", "M")
```

```python
search_long = search_month.melt(id_vars = "date",
    value_vars = ["cold", "flu", "pneumonia", "coronavirus"])
```

```python
locationWeather_long = locationWeather[['date', 'province', 'avg_temp', "confirmed", "deceased", "min_temp", "max_temp"]].melt(id_vars = ["date", "province"],
                                                                             value_vars = ["avg_temp", "confirmed", "deceased"])
```

```python
base = alt.Chart(locationWeather).encode(
    alt.X('date:T', axis=alt.Axis(title=None))
)

area = base.mark_area(opacity=0.3, color='#57A44C').encode(
    alt.Y('average(max_temp)',
          axis=alt.Axis(title='Daily Temperature Range (C)', titleColor='#57A44C')),
    alt.Y2('average(min_temp)')
)

line = base.mark_line(stroke='#5276A7', interpolate='monotone').encode(
    alt.Y('average(confirmed)',
          axis=alt.Axis(title='Confirmed Cases', titleColor='#5276A7'))
)



alt.concat(*(
    alt.layer(area, line, title=f"{val}").transform_filter(alt.datum.province == val).resolve_scale(y='independent',x='independent')
    for val in locationWeather.province.unique()
))
```
